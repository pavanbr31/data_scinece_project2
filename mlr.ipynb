```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mlr.csv")
print(df)
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(x)
print(y)
```

```python
z=df.isna()
print(z)
```

```python
df.isna().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers =[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x = ct.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.2,random_state=0)
print(y_te)
```

```python
#create mlr model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_tr,y_tr)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_pred = lr.predict(x_te)
mse = mean_squared_error(y_te,y_pred)
r2 = r2_score(y_te,y_pred)
print(mse)
print(r2)
```

#BACKWARD ELIMINATION PROCESS

```python
x_temp = x[:, 1:]
const = np.ones((50,1))
x_temp = np.append(arr = const,values = x,axis = 1)
print(x_temp)
```

```python
import statsmodels.api as sm
x_opt = np.array(x_temp[:,[0,1,2,3,4]],dtype=float)
stats = sm.OLS(endog=y,exog=x_opt).fit()
print(stats.summary())
```

```python
xo_te,xo_tr,yo_te,yo_tr=train_test_split(x_opt,y,test_size=0.2,random_state=0)
lr_opt = LinearRegression()
lr_opt.fit(xo_tr,yo_tr)
```

```python
yo_pred = lr_opt.predict(xo_te)
mse = mean_squared_error(yo_te,yo_pred)
r2 = r2_score(yo_te,yo_pred)
print(mse)
print(r2)
```
